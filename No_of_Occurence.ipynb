```python
import time
import random



```

This cell imports the `time` module, which provides functions for working with time, such as `time.sleep()` for pausing execution. It also imports the `random` module, which is used for generating pseudo-random numbers, like `random.randint()` and `random.uniform()`.

## Packet Arrival WITHOUT Delay

This markdown cell introduces the section that demonstrates packet arrival without any simulated network delay.

```python
print("=== Packet Stream Without Delay ===")

for packet_id in range(1, 11):
    packet_size = random.randint(64, 1500)  # Packet size in bytes

    print(f"Packet {packet_id} arrived | Size: {packet_size} bytes")

print("\n")
```

This cell simulates a stream of 10 packets arriving without any delay. It prints a message for each packet, indicating its `packet_id` and a randomly generated `packet_size` (between 64 and 1500 bytes).

## Packet Arrival WITH Delay

This markdown cell introduces the section that demonstrates packet arrival with simulated network delay.

```python
print("=== Packet Stream With Delay ===")

for packet_id in range(1, 11):
    packet_size = random.randint(64, 1500)  # Packet size in bytes

    # Simulated network delay (0.5 to 2 seconds)
    delay = random.uniform(0.5, 2.0)

    time.sleep(delay)

    print(
        f"Packet {packet_id} arrived | "
        f"Size: {packet_size} bytes | "
        f"Delay: {delay:.2f} sec"
    )

print("\nTransmission Complete.")
```

This cell simulates a stream of 10 packets arriving with a random delay. For each packet, it generates a `packet_size` and a `delay` (between 0.5 and 2.0 seconds). The `time.sleep(delay)` function pauses the execution for the simulated delay duration before printing the packet arrival information, including the `packet_id`, `packet_size`, and the simulated `delay`.
