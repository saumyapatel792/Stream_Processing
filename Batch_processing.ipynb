# Batch Processing Case Study for Stream Processing

```python
import pandas as pd
```

This cell imports the `pandas` library, which is essential for data manipulation and analysis in Python. It's aliased as `pd` for convenience.

```python
# Read CSV file
df = pd.read_csv("student_marks.csv")
df.head()
```

This cell reads the `student_marks.csv` file into a pandas DataFrame called `df`. The `df.head()` function displays the first few rows of the DataFrame, providing a quick look at the data structure and content.

```python
# Function to assign grades
def assign_grade(mark):
    if mark >= 90:
        return "A"
    elif mark >= 80:
        return "B"
    elif mark >= 70:
        return "C"
    else:
        return "D"
```

This cell defines a Python function `assign_grade` that takes a student's mark as input and returns a letter grade (A, B, C, or D) based on a predefined grading scale. This function will be used to categorize students' performance.

```python
# Add Grade column
df["Grade"] = df["Marks"].apply(assign_grade)
```

Here, the `assign_grade` function is applied to the 'Marks' column of the DataFrame `df`. A new column named 'Grade' is created, storing the corresponding letter grade for each student based on their marks.

```python
# Batch Processing Summary
average_marks = df["Marks"].mean()
highest_marks = df["Marks"].max()
lowest_marks = df["Marks"].min()
```

This cell calculates key descriptive statistics for the 'Marks' column: the average marks, the highest marks, and the lowest marks among all students. These statistics provide a quick summary of the students' performance distribution.

```python
# Display processed dataset
print("Processed Dataset:")
print(df)

print("\nBatch Processing Summary")
print("------------------------")
print("Average Marks:", round(average_marks, 2))
print("Highest Marks:", highest_marks)
print("Lowest Marks:", lowest_marks)
```

This cell prints the processed DataFrame, including the newly added 'Grade' column, and then displays the batch processing summary (average, highest, and lowest marks). This output allows for a comprehensive review of the data after processing and a quick glance at the overall performance.

```python
# Save output
df.to_csv("processed_student_marks.csv", index=False)

print("\nProcessed file saved as processed_student_marks.csv")
```

Finally, this cell saves the modified DataFrame `df` (which now includes the 'Grade' column) to a new CSV file named `processed_student_marks.csv`. The `index=False` argument prevents pandas from writing the DataFrame index as a column in the CSV file, and a confirmation message is printed.
