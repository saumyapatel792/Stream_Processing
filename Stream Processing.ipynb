## Random Data Stream Generation and Visualization

```python
import random
import time
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options for better visualization of DataFrames
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

### Data Generation Parameters

We will simulate a stream of 100 data points, with each data point generated 15 seconds after the previous one. Each data point will be a random integer between 1 and 1000.

```python
# Parameters for the data stream
num_data_points = 100 # Number of data points to generate
interval_seconds = 15 # Time interval between data points in seconds

data_stream = []
current_time = datetime.now() # Start time for the stream

print(f"Generating {num_data_points} data points with a {interval_seconds}-second interval...")

for i in range(num_data_points):
    # Generate a random integer between 1 and 1000
    random_value = random.randint(1, 1000)

    # Store the timestamp and the generated value
    data_stream.append({
        'timestamp': current_time,
        'value': random_value
    })

    # Increment time for the next data point
    current_time += timedelta(seconds=interval_seconds)

print("Data generation complete.")
```

### Process and Display Data

Convert the generated data into a pandas DataFrame for easier manipulation and display the first few rows.

```python
df = pd.DataFrame(data_stream)

# Ensure 'timestamp' column is of datetime type
df['timestamp'] = pd.to_datetime(df['timestamp'])

print("Generated Data:")
display(df.head())
```

### Visualize the Data Stream

We will create two types of visualizations:
1.  **Line Plot**: To show how the random values change over time.
2.  **Histogram**: To show the distribution of the generated random values.

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='timestamp', y='value', data=df)
plt.title('Random Data Stream Over Time (15-sec intervals)')
plt.xlabel('Timestamp')
plt.ylabel('Random Value (1-1000)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['value'], bins=20, kde=True)
plt.title('Distribution of Random Data Values')
plt.xlabel('Random Value (1-1000)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
